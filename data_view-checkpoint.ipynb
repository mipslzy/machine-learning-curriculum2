```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
```

```python
# 获取相关数据
train = pd.read_csv("train1_data.csv")
test = pd.read_csv("test_data.csv")
```

```python
# 训练集正负样例数据统计
P = train[train["is_default"] == 1]
N = train[train["is_default"] == 0]
P_num = len(P)
N_num = len(N)
P_num,N_num
```

```python
# 训练集数据正负样例分布情况可视化
plt.bar(["正例","负例"],[P_num,N_num])
plt.xlabel("类别")
plt.ylabel("数量")
plt.title("样例分布情况")
plt.show()
```

```python
# 训练集数据借款人在贷款评分分布情况数据
scoring_high = train['scoring_high']
scoring_high
```

```python
# 训练集数据借款人在贷款评分分布情况数据可视化
plt.boxplot(scoring_high)
plt.ylabel("贷款评分")
plt.title("贷款评分分布情况")
plt.show()
```

```python
# 公开可用策略数据
policy_count = train['policy_code'].value_counts()
policy_count
```

```python
# 匿名特征分布
af = train[['f0','f1','f2','f3','f4']]
plt.xlabel('样例')
plt.ylabel('取值')
plt.title('匿名特征分布')
plt.legend(loc='upper right')
plt.show()
```

```python
# 初始列表状态贷款数量占比
ilsc = train['initial_list_status'].value_counts()
ilsc
```

```python
ilsc.plot(kind='pie', figsize=(8, 8), autopct='%1.1f%%')
plt.title('初始列表状态贷款数量占比')
plt.ylabel('')
plt.show()
```

```python
# 借贷人提供贷款名称分布图
title_counts = train['title'].value_counts()
top_10 = title_counts.head(10)
top_10
```

```python
top_10.plot(kind = 'bar',figsize = (10,6))
plt.xlabel("标题")
plt.ylabel("出现频率")
plt.title("前十标题的出现频率")
plt.show()
```

```python
# 年收入的分布情况

 # 设置区间范围和步长
bin_range = (0,100000,5000)

# 使用cut函数将数据分组并计数
loan_counts = pd.cut(train['total_loan'],bins = bin_range,include_lowest = True).value_counts().sort_index()
new = sorted(loan_counts)
# 绘制折线图
plt.figure(figsize=(12,6))
plt.plot(new.index.astype(str),new.values,marker='o',linestyle='-',linewidth=2)
plt.xlabel('贷款范围')
plt.ylabel('钱数')
plt.title('借贷分布情况')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
